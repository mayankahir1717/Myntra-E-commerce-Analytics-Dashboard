```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Myntra_dataset.csv')
df.head()
```

```python
# Q1. What is the shape of the dataset?

df.shape
```

```python
# Q2. Show basic info — column names, data types, null counts

df.info()
```

```python
# Q3. Show summary statistics of numeric columns
df.describe()
```

```python
# Q4. How many null values are in each column?
df.isnull().sum()
```

```python
# Q5. How many orders were Delivered, Cancelled, Returned?
df['Order_Status'].value_counts()
```

```python
# Q6. Plot a bar chart of orders by Order Status
df['Order_Status'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Orders by Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Q7. What is the total revenue by Region?
df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
```

```python
# Q8. Plot total revenue by Region


df.groupby('Region')['Revenue'].sum().sort_values().plot(kind='barh', color='coral')
plt.title('Revenue by Region')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# Q9. Which are the Top 10 Brands by Revenue?
df.groupby('Brand')['Revenue'].sum().nlargest(10).sort_values().plot(kind='barh', color='teal')
plt.title('Top 10 Brands by Revenue')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# Q10. What is the average Order Value by Gender?
df.groupby('Gender')['Order_Value'].mean()
```

```python
# Q11. Plot orders by Payment Method
df['Payment_Method'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Orders by Payment Method')
plt.ylabel('')
plt.show()
```

```python
# Q12. Fill nulls in Cancellation_Reason, Return_Reason, Delivery_Reason
# These are null because order wasn't cancelled/returned — fill with 'N/A'
df['Cancellation_Reason'] = df['Cancellation_Reason'].fillna('N/A')
df['Return_Reason'] = df['Return_Reason'].fillna('N/A')
df['Delivery_Reason'] = df['Delivery_Reason'].fillna('N/A')

df[['Cancellation_Reason', 'Return_Reason', 'Delivery_Reason']].isnull().sum()
```

```python
# Q13. Check for duplicate rows
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Q14. Drop duplicate rows (if any)
df = df.drop_duplicates()
print("After dropping duplicates:", df.shape)
```

```python
# First convert Order_Date to datetime 
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

# Q15. Extract useful date parts from Order_Date
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
df['Month_Name'] = df['Order_Date'].dt.strftime('%B')
df['Day_of_Week'] = df['Order_Date'].dt.day_name()

df[['Order_Date', 'Year', 'Month', 'Month_Name', 'Day_of_Week']].head()
```

```python
# Q16. Save cleaned dataframe to new Excel file
df.to_excel('Myntra_Clean_dataset.xlsx', index=False)
print("Saved successfully!")
```

```python
# Q17. Final summary — shape, nulls, dtypes check
print("Shape:", df.shape)
print("\nNull Values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
```
